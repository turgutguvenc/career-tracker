```python
import psycopg2
from datetime import datetime
from serpapi import GoogleSearch
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from .env file in the parent directory
load_dotenv(find_dotenv())

# PostgreSQL connection parameters from environment variables
DB_HOST = os.getenv("DB_HOST")
DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
SERPAPI_API_KEY = os.getenv("SERPAPI_API_KEY")

# SerpApi configuration
serpapi_params_jobs = {
    "engine": "google_jobs",
    "q": "Data Analyst",
    "location": "Boston, Massachusetts, United States",
    "hl": "en",
    "api_key": SERPAPI_API_KEY,
}

params_event = {
    "engine": "google_events",
    "q": "job fairs in Boston",
    "hl": "en",
    "gl": "us",
    "api_key": SERPAPI_API_KEY,
}

def collect_serpapi_job_data(serpapi_params, cur, conn):
    search = GoogleSearch(serpapi_params)
    results = search.get_dict()
    jobs_results = results.get("jobs_results", [])
    for job in jobs_results:
        title = job.get("title")
        company_name = job.get("company_name")
        location = job.get("location")
        date_posted = job.get("detected_extensions", {}).get("posted_at")
        
        # Ensure date_posted is not None
        if date_posted:
            try:
                date_posted = datetime.strptime(date_posted, "%Y-%m-%d")
            except ValueError:
                date_posted = datetime.now()  # Fallback to current date if parsing fails
        else:
            date_posted = datetime.now()
        
        url = job.get('related_links', [{}])[0].get('link', '')
        description = job.get("description", "No description provided.")

        # Insert job data into PostgreSQL
        cur.execute("""
            INSERT INTO job_postings (title, company_name, location, date, url, description)
            VALUES (%s, %s, %s, %s, %s, %s)
        """, (title, company_name, location, date_posted, url, description))
        conn.commit()
    print(f"Inserted {len(jobs_results)} jobs into the database.")

def collect_serpapi_event_data(serpapi_params, cur, conn):
    search = GoogleSearch(serpapi_params)
    results = search.get_dict()
    events_results = results.get("events_results", [])
    for event in events_results:
        name = event.get('title')
        
        # Parse the date string
        date = event.get('date', {}).get('start_date', '')
        
        # Join address parts into a single string
        location = ', '.join(event.get('address', []))
        
        url = event.get('link', '')
        img_url = event.get('image', '') 

        # Insert event data into PostgreSQL
        cur.execute("""
            INSERT INTO events (name, date, location, url, img_url)
            VALUES (%s, %s, %s, %s, %s)
        """, (name, date, location, url, img_url))
        conn.commit()
    print(f"Inserted {len(events_results)} events into the database.")

def main():
    """
    - Establishes connection with the job_events_db database and gets
    cursor to it.
    
    - Collects job and event data and inserts it into the database.
    
    - Finally, closes the connection.
    """
    # PostgreSQL connection
    conn = psycopg2.connect(
        host=DB_HOST,
        dbname=DB_NAME,
        user=DB_USER,
        password=DB_PASSWORD
    )
    cur = conn.cursor()
    # Collect and insert job data
    collect_serpapi_job_data(serpapi_params_jobs, cur, conn)
    collect_serpapi_event_data(params_event, cur, conn)
    conn.close()

if __name__ == "__main__":
    main()
```

```python
import psycopg2
from datetime import datetime
from serpapi import GoogleSearch
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from .env file in the parent directory
load_dotenv(find_dotenv())

# PostgreSQL connection parameters from environment variables
DB_HOST = os.getenv("DB_HOST")
DB_NAME = os.getenv("DB_NAME")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
SERPAPI_API_KEY = os.getenv("SERPAPI_API_KEY")

# SerpApi configuration
serpapi_params_jobs = {
    "engine": "google_jobs",
    "q": "Data Analyst",
    "location": "Boston, Massachusetts, United States",
    "hl": "en",
    "api_key": SERPAPI_API_KEY,
}

serpapi_params_events = {
    "engine": "google_events",
    "q": "job fairs in Boston",
    "hl": "en",
    "gl": "us",
    "api_key": SERPAPI_API_KEY,
}

def collect_serpapi_job_data(serpapi_params, cur, conn):
    jobs_results = []
    start = 0
    while True:
        serpapi_params['start'] = start
        search = GoogleSearch(serpapi_params)
        results = search.get_dict()
        new_jobs_results = results.get("jobs_results", [])
        if not new_jobs_results:
            break
        jobs_results.extend(new_jobs_results)
        start += len(new_jobs_results)

        #I only want to see 100 results each work
        if len(jobs_results ) >= 100:
            break
    for job in jobs_results:
        title = job.get("title")
        company_name = job.get("company_name")
        location = job.get("location")
        date_posted = job.get("detected_extensions", {}).get("posted_at")
        
        # Ensure date_posted is not None
        if date_posted:
            try:
                date_posted = datetime.strptime(date_posted, "%Y-%m-%d")
            except ValueError:
                date_posted = datetime.now()  # Fallback to current date if parsing fails
        else:
            date_posted = datetime.now()
        
        url = job.get('related_links', [{}])[0].get('link', '')
        description = job.get("description", "No description provided.")

        # Insert job data into PostgreSQL
        cur.execute("""
            INSERT INTO job_postings (title, company_name, location, date, url, description)
            VALUES (%s, %s, %s, %s, %s, %s)
        """, (title, company_name, location, date_posted, url, description))
        conn.commit()
    print(f"Inserted {len(jobs_results)} jobs into the database.")

def collect_serpapi_event_data(serpapi_params, cur, conn):
    events_results = []
    start = 0
    while True:
        serpapi_params['start'] = start
        search = GoogleSearch(serpapi_params)
        results = search.get_dict()
        new_events_results = results.get("events_results", [])
        if not new_events_results:
            break
        events_results.extend(new_events_results)
        start += len(new_events_results)
         #I only want to see 100 results each work
        if len(events_results) >= 100:
            break
    for event in events_results:
        name = event.get('title')
        
        # Use the provided date string directly
        date = event.get('date', {}).get('start_date', '')
        
        # Join address parts into a single string
        location = ', '.join(event.get('address', []))
        
        url = event.get('link', '')
        img_url = event.get('image', '') 

        # Insert event data into PostgreSQL
        cur.execute("""
            INSERT INTO events (name, date, location, url, img_url)
            VALUES (%s, %s, %s, %s, %s)
        """, (name, date, location, url, img_url))
        conn.commit()
    print(f"Inserted {len(events_results)} events into the database.")

def main():
    """
    - Establishes connection with the job_events_db database and gets
    cursor to it.
    
    - Collects job and event data and inserts it into the database.
    
    - Finally, closes the connection.
    """
    # PostgreSQL connection
    conn = psycopg2.connect(
        host=DB_HOST,
        dbname=DB_NAME,
        user=DB_USER,
        password=DB_PASSWORD
    )
    cur = conn.cursor()
    # Collect and insert job data
    collect_serpapi_job_data(serpapi_params_jobs, cur, conn)
    collect_serpapi_event_data(serpapi_params_events, cur, conn)
    conn.close()

if __name__ == "__main__":
    main()
```
